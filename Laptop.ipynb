Numero 6

```python
def filter(datafr: pd.DataFrame, class_name: str) -> pd.DataFrame:
    """Filters DataFrame by class

    Args:
        datafr (pd.DataFrame): DataFrame
        class_name (str): Cat or Dog

    Returns:
        pd.DataFrame: filtered DataFrame
    """
    return datafr.loc[datafr["class"] == class_name]


new_frame = filter(datafr, 'Cat')
print(new_frame)
```

Numero 7

```python
def advanced_filter(datafr: pd.DataFrame, class_name: str, max_width: int, max_height: int) -> pd.DataFrame:
    """Returns new dataframe with pictures that meet the requierements

    Args:
        datafr (pd.DataFrame): DataFrame
        class_name (str): Cat or Dog
        max_width (int): max allowed width for filter
        max_height (int): max allowed height for filter

    Returns:
        pd.DataFrame: New filtered dataframe
    """
    return datafr.loc[(datafr["class"] == class_name) & (datafr["width"] <= max_width) & (datafr["height"] <= max_height)]


new_frame = advanced_filter(datafr, 'Cat', 1800, 6000)
print(new_frame)
```

Numero 9

```python
def get_histo(datafr: pd.DataFrame, class_name: str):
    """gets histogramm for each color level of randomly chosen image

    Args:
        datafr (pd.DataFrame): DataFrame
        class_name (str): Cat or Dog

    Returns:
        tuple[List, MatLike]: color[] - list of 3 arrays, image - MatLike
    """

    image_path = f'dataset/{class_name}s/{
        str(random.randint(0, 1000)).zfill(4)}.jpg'
    image = cv2.imread(image_path)

    color = []
    for i in range(0, 3):
        histo = cv2.calcHist([image], [i], None, [256], [0, 256])
        color.append(histo)
    return color, image


print(get_histo(datafr, 'Cat'))
```

Numero 10

```python
def draw_hist(df: pd.DataFrame, class_name: str) -> None:
    """draws a histogramm using data from get_histo

    Args:
        df (pd.DataFrame): DataFrame
        class_name (str): Cat or Dog
    """
    spect, image = get_histo(datafr, class_name)
    plt.subplot(3, 1, 1)
    plt.imshow(image)
    
    plt.subplot(2, 1, 2)
    plt.plot(spect[0], color='b')
    plt.plot(spect[1], color='g')
    plt.plot(spect[2], color='r')
    
    plt.title('Histogramma')
    plt.xlabel('Color value')
    plt.ylabel('How many much')
    plt.legend(['Blue', 'Green', 'Red'])

    plt.show()


draw_hist(datafr, 'Cat')
```
